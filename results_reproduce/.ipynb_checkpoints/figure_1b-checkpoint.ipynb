#### Read quality plot

This script takes fastq file as input to calculate read-level base quality.
The fastq files used in this script are publicly availbale at SRA database under the accession number [SRP166020](https://www.ncbi.nlm.nih.gov/sra/?term=SRP166020)

```python
#read quality plot

import traceback
import matplotlib.pyplot as plt
from plotnine import *
import pandas as pd
import re
import numpy as np
from statsmodels import robust

black = '#222222'
gray = '#666666'
red = '#FF3333'
green = '#66CC00'
blue = '#3333FF'
purple = '#9933FF'
orange = '#FF8000'
yellow = '#FFFF33'


def get_mean_quality(file):
    """
    Retrieves the mean quality scores from fastq file.

    Args:
        file (str): Path to the file containing sequencing data.

    Returns:
        list: List of mean quality scores.

    """
    count=0
    mean_quality_list=[]
    next=""
    with open(file) as f:
        for line in f:
            line=line.rstrip()
            if "runid" in line:
                next="sequence"
                continue
            if  line=="+":
                next="quality"
                continue
            if next=="quality":
                base_quality_string=line
                base_quality_list=[ord(char) for char in base_quality_string]
                mean_quality=np.mean(base_quality_list)-33
                mean_quality_list.append(mean_quality)
                count+=1 
            if count>10000:
                break
    return mean_quality_list


X=[]
Y=[]

file="/home/wuyou/Projects/paper/ELIGOS_normalA/IVT_normalA.fastq"
mean_quality_list=get_mean_quality(file)
X.extend(mean_quality_list)
Y.extend(["unmod"]*len(mean_quality_list))

file="/home/wuyou/Projects/paper/ELIGOS_m1A/IVT_m1A.fastq"
mean_quality_list=get_mean_quality(file)
X.extend(mean_quality_list)
Y.extend(["m1A"]*len(mean_quality_list))

file="/home/wuyou/Projects/paper/ELIGOS_m6A/IVT_m6A.fastq"
mean_quality_list=get_mean_quality(file)
X.extend(mean_quality_list)
Y.extend(["m6A"]*len(mean_quality_list))

file="/home/wuyou/Projects/paper/ELIGOS_m5C/IVT_m5C.fastq"
mean_quality_list=get_mean_quality(file)
X.extend(mean_quality_list)
Y.extend(["m5C"]*len(mean_quality_list))

file="/home/wuyou/Projects/paper/ELIGOS_hm5C/IVT_hm5C.fastq"
mean_quality_list=get_mean_quality(file)
X.extend(mean_quality_list)
Y.extend(["hm5C"]*len(mean_quality_list))

file="/home/wuyou/Projects/paper/ELIGOS_m7G/IVT_m7G.fastq"
mean_quality_list=get_mean_quality(file)
X.extend(mean_quality_list)
Y.extend(["m7G"]*len(mean_quality_list))

file="/home/wuyou/Projects/paper/ELIGOS_psU/IVT_psU.fastq"
mean_quality_list=get_mean_quality(file)
X.extend(mean_quality_list)
Y.extend([""]*len(mean_quality_list))


data=pd.DataFrame(X)
#data["label"]=Y
data.columns=["quality"]
data["modification"]=Y

data.to_csv("data/event_level_features_read_quality_m1A_m6A_m5C_hm5C_m7G_psU.csv")

print("plotting")
p1=(ggplot()
    +geom_density(data,aes(x="quality",color='modification', fill='modification'),alpha=0.2)
    +theme(
        figure_size=[3.04,1.22],
        axis_text=element_text(size=6,family="Arial",color="black"),
        axis_title_x=element_text(size=6,family="Arial",color="black"),
        axis_title_y=element_text(size=6,family="Arial",color="black"),
        panel_background=element_rect(fill=gray, alpha=0),
        panel_grid_major=element_line(size=0.3, alpha=0.3,color=black),
        panel_grid_minor=element_line(size=0.3, alpha=0.3,color=black),
        panel_border=element_rect(color=black, size=1),
        legend_title = element_text(size=6,alpha=0), #change legend title font size
        legend_text = element_text(size=6),
        legend_background=element_rect(size=0.5,alpha=0),
        #legend_position=(0.25,0.55), 
        legend_key_size=4) #change legend text font size))
   +xlab("Quality")
   +ylab("Density")
   )

print(p1)
"""
p2=(ggplot()+geom_line(data_mod,aes(x="x",y="value",group="variable"),alpha=0.05,color="red")
    +geom_line(data_unmod,aes(x="x",y="value",group="variable"),alpha=0.05,color="black")
    +theme(panel_background=element_rect(fill=gray, alpha=0),
            panel_grid_major=element_line(size=0.3, alpha=0.3,color=black),
            panel_grid_minor=element_line(size=0.3, alpha=0.3,color=black),
            panel_border=element_rect(color=black, size=1),
            legend_position = "none",
          figure_size=[6,2.5])+labs(y = "normalized signal")   
)
p2
"""
p1.save('figure/figure_1b_event_level_features_base_quality.pdf')

```
