```python

import matplotlib.pyplot as plt
from plotnine import *
import pandas as pd
import numpy as np

#
m1A="/home/wuyou/Projects/Os_AD/data/predict/false_negitive_unmod_m1A.predict.txt"
m6A="/home/wuyou/Projects/Os_AD/data/predict/false_negitive_unmod_m6A.predict.txt"
m5C="/home/wuyou/Projects/Os_AD/data/predict/false_negitive_unmod_m5C.predict.txt"
hm5C="/home/wuyou/Projects/Os_AD/data/predict/false_negitive_unmod_hm5C.predict_2.27.txt"
m7G="/home/wuyou/Projects/Os_AD/data/predict/false_negitive_unmod_m7G.predict.txt"
psU="/home/wuyou/Projects/Os_AD/data/predict/false_negitive_unmod_psU.predict.txt"

mod_rates=[]
mods=[]

cut_off=0.95

#m1A
count_dict={}
with open(m1A) as f:
    for line in f:
        gene=line.split("\t")[0]
        if gene not in count_dict:
            count_dict[gene]=[0,0]
        probability=float(line.split("\t")[5])
        if probability>cut_off:
            count_dict[gene][0]+=1
        count_dict[gene][1]+=1
for gene in count_dict:
    mod_rate=count_dict[gene][0]/count_dict[gene][1]
    mod_rates.append(mod_rate)
    mods.append("m1A")
        
#m6A
count_dict={}
with open(m6A) as f:
    for line in f:
        gene=line.split("\t")[0]
        if gene not in count_dict:
            count_dict[gene]=[0,0]
        probability=float(line.split("\t")[5])
        if probability>cut_off:
            count_dict[gene][0]+=1
        count_dict[gene][1]+=1
for gene in count_dict:
    mod_rate=count_dict[gene][0]/count_dict[gene][1]
    mod_rates.append(mod_rate)
    mods.append("m6A")        

#m5C
count_dict={}
with open(m5C) as f:
    for line in f:
        gene=line.split("\t")[0]
        if gene not in count_dict:
            count_dict[gene]=[0,0]
        probability=float(line.split("\t")[5])
        if probability>cut_off:
            count_dict[gene][0]+=1
        count_dict[gene][1]+=1
for gene in count_dict:
    mod_rate=count_dict[gene][0]/count_dict[gene][1]
    mod_rates.append(mod_rate)
    mods.append("m5C")  

#hm5C
count_dict={}
with open(hm5C) as f:
    for line in f:
        gene=line.split("\t")[0]
        if gene not in count_dict:
            count_dict[gene]=[0,0]
        probability=float(line.split("\t")[5])
        if probability>cut_off:
            count_dict[gene][0]+=1
        count_dict[gene][1]+=1
for gene in count_dict:
    mod_rate=count_dict[gene][0]/count_dict[gene][1]
    mod_rates.append(mod_rate)
    mods.append("hm5C")  
    
#m7G
count_dict={}
with open(m7G) as f:
    for line in f:
        gene=line.split("\t")[0]
        if gene not in count_dict:
            count_dict[gene]=[0,0]
        probability=float(line.split("\t")[5])
        if probability>cut_off:
            count_dict[gene][0]+=1
        count_dict[gene][1]+=1
for gene in count_dict:
    mod_rate=count_dict[gene][0]/count_dict[gene][1]
    mod_rates.append(mod_rate)
    mods.append("m7G")  
    
#psU
count_dict={}
with open(psU) as f:
    for line in f:
        gene=line.split("\t")[0]
        if gene not in count_dict:
            count_dict[gene]=[0,0]
        probability=float(line.split("\t")[5])
        if probability>cut_off:
            count_dict[gene][0]+=1
        count_dict[gene][1]+=1
for gene in count_dict:
    mod_rate=count_dict[gene][0]/count_dict[gene][1]
    mod_rates.append(mod_rate)
    mods.append("Î¨")  

    
black = '#222222'
gray = '#666666'
red = '#FF3333'
green = '#66CC00'
blue = '#3333FF'
purple = '#9933FF'
orange = '#FF8000'
yellow = '#FFFF33'
c1="#F8766D"
c2="#00BA38" 
c3="#619CFF"



data=pd.DataFrame({"false positve rate":mod_rates,"modification":mods})
#data['sample'] = pd.Categorical(data['sample'], categories=pd.unique(data['sample']))  #reorder legend

p=(
    ggplot()
    +geom_boxplot(data,aes(y="false positve rate",x="modification",fill="modification"),alpha=0.7,outlier_shape = "")
     #+scale_x_continuous(breaks = [0,100,200,300],labels =["               5'UTR", "                CDS", "                3'UTR",""])
    +theme(panel_background=element_rect(fill=gray, alpha=0),
            strip_text_x = element_text(size = 6),            #facet font size
            legend_position = "none",
            panel_grid_major=element_line(size=0.3, alpha=0.3,color=black),
            panel_grid_minor=element_line(size=0.3, alpha=0.3,color=black),
            panel_border=element_rect(color=black, size=1),
            axis_text=element_text(size=6,family="Arial",color="black"),
            axis_title_x=element_text(size=6,family="Arial",color="black"),
            axis_title_y=element_text(size=6,family="Arial",color="black"),
             
           
            #figure_size=[1.82,1.33],
            figure_size=[2.2,1.33],
            legend_title = element_text(size=6), #change legend title font size
            legend_text = element_text(size=6),
            legend_background=element_rect(size=0.5,alpha=0),
            #legend_position=(0.45,0.4), 
            legend_key_size=12) #change legend text font size)
    #+ geom_vline(xintercept=100,linetype="dashed",alpha=0.5)
    #+ geom_vline(xintercept=200,linetype="dashed",alpha=0.5)
    #+ylim([0,0.20])
    #+scale_fill_manual(values=[c1, c2, c3]) 
    +labs(x="",y="False positive")
)

print(p)
p.save('figure/figure_4d_mod_false_positive_evaluation.pdf')
```
